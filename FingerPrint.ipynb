```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('finger.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
oneDImg = img[:,:,0]
oneDImg = oneDImg.reshape((-1,))
print(oneDImg.shape)
print(oneDImg)
```

```python
for i in range(oneDImg.shape[0]):
    if oneDImg[i] == 255:
        oneDImg[i] = 0
    else:
        oneDImg[i] = 1
print(oneDImg)
print(oneDImg.shape)
```

```python
for i in range(oneDImg.shape[0]):
    print(oneDImg[i],end=" ")
```

```python
with open("imageMatrix.txt",'w') as f:
    for i in oneDImg:
        f.write(str(i))
    f.write('+')
```
